```python
df.drop(['url', 'address', 'phone'], axis=1, inplace=True)
df.columns.tolist()
```

```python
df.rename(columns={
    'rate': 'ratings',
    'approx_cost(for two people)': 'approx_cost_for_two',
    'listed_in(type)': 'listed_in_type',
    'listed_in(city)': 'listed_in_city'
}, inplace=True)

df.columns.tolist()
```

```python
df['ratings'] = df['ratings'].str.replace('/5', '', regex=False)
df['ratings'] = df['ratings'].replace('NEW', np.nan)
df['ratings'] = df['ratings'].replace('-', np.nan)

df['ratings'].value_counts(dropna=False)
```

```python
# Can't fabricate a rating — a missing/unrated restaurant is dropped,
# not imputed, since guessing a rating would bias every downstream average.
df = df.dropna(subset=['ratings'])
df['ratings'] = pd.to_numeric(df['ratings'])
df['ratings'].dtype
```

```python
df['approx_cost_for_two'].unique()
```

```python
df['approx_cost_for_two'] = df['approx_cost_for_two'].str.replace(',', '', regex=False)
df['approx_cost_for_two'] = pd.to_numeric(df['approx_cost_for_two'])
df['approx_cost_for_two'].isnull().sum()
```

```python
median_cost = df['approx_cost_for_two'].median()
df['approx_cost_for_two'] = df['approx_cost_for_two'].fillna(median_cost)

print(f"Filled with median cost: {median_cost}")
df['approx_cost_for_two'].isnull().sum()
```

```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower) | (data[column] > upper)]
    print(f"{column}: {len(outliers)} outliers ({len(outliers)/len(data)*100:.1f}%), "
          f"bounds [{lower:.1f}, {upper:.1f}]")
    return outliers

votes_outliers = detect_outliers_iqr(df, 'votes')
cost_outliers = detect_outliers_iqr(df, 'approx_cost_for_two')
```

```python
restaurant_rating = (
    df.groupby('name')
      .agg(Average_Rating=('ratings', 'mean'), Number_of_Outlets=('name', 'count'))
)

restaurant_rating = restaurant_rating[restaurant_rating['Number_of_Outlets'] >= 20]
restaurant_rating = restaurant_rating.sort_values(by='Average_Rating', ascending=False)

restaurant_rating.head(10)
```
