```python
# restaurant outlets with highest average ratings with atleast 20 outlets

restaurant_rating = (
    df.groupby('name')
      .agg(
          Average_Rating=('ratings', 'mean'),
          Number_of_Outlets=('name', 'count')
      )
)
```

```python
restaurant_rating = restaurant_rating[
    restaurant_rating['Number_of_Outlets'] >= 20
]
```

```python
restaurant_rating = restaurant_rating.sort_values(
    by='Average_Rating',
    ascending=False
)

restaurant_rating.head(10)
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    data=restaurant_rating.head(10),
    x='Average_Rating',
    y=restaurant_rating.head(10).index
)

plt.title('Top Rated Restaurant Chains (Min 20 Outlets)')
plt.xlabel('Average Rating')
plt.ylabel('Restaurant Chain')

plt.show()
```
